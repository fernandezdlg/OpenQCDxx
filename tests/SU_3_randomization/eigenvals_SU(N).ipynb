```python
# Target distribution (from paper: https://arxiv.org/pdf/2401.09045)
N = 3

x = np.linspace(-np.pi, np.pi, 1000)
y = (N/(2*np.pi)) * ( 1 - (-1)**N * (2/N) * np.cos(x*N) )

plt.plot(x, y)
plt.xlabel('Eigenvalues')
plt.ylabel('Target Density')

# Plot horizontal lines for max and min values of y
plt.axhline(y.max(), color='r', linestyle='--', label='Max y')
plt.axhline(y.min(), color='b', linestyle='--', label='Min y')

plt.show()
```
